```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices, columns=["month", "price_pln"])
df= pd.DataFrame(df, columns=["month", "price_pln", "price_usd"])
```

```python
price_usd=df['price_pln'].apply(lambda price_pln: price_pln/4)
df['price_usd']=price_usd
```

```python
df=df.set_index('month')
plt.plot(df.index, df['price_usd'], linestyle='--', color="red")
plt.title("Price of goods (USD)")
plt.xlabel("month")
plt.grid(True)
```
